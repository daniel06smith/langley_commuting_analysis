```python
# Get our data; import similar_cities dataset
similar_cities = pd.read_csv("../data/processed/similar_cities.csv")
chilliwack_data = similar_cities[similar_cities["Geography"] == "Chilliwack"]
nanaimo_data = similar_cities[similar_cities["Geography"] == "Nanaimo"]
langley_data = similar_cities[similar_cities["Geography"] == "Langley"]
kamloops_data = similar_cities[similar_cities["Geography"] == "Kamloops"]

# Renaming to transit_cities for simplicity
transit_cities = pd.concat([langley_data, nanaimo_data, chilliwack_data, kamloops_data])
print(transit_cities)
```

Since we are interested in public transportation-related information in this case study, let's filter our columns to just contain commuting data specific to public transit.

```python
# Filter the public transit related data
public_transit = []
for column in transit_cities.columns:
    if column.startswith("Public_Transit_") and column != "Public_Transit_Avg_Duration":
        public_transit.append(column)
print(public_transit)

# Let's make it look good for the graph labels
pretty_public_transit = [column.replace("Public_Transit_", "").replace("_"," ") for column in public_transit]
print(pretty_public_transit)
```

Next let's extract the relevant data for each column for each city.

```python
# Extract commuting period data for Langley and Nanaimo
langley_values = langley_data[public_transit].to_numpy().flatten()
nanaimo_values = nanaimo_data[public_transit].to_numpy().flatten()
kamloops_values = kamloops_data[public_transit].to_numpy().flatten()
chilliwack_values = chilliwack_data[public_transit].to_numpy().flatten()

# Verify these values
print(langley_values)
print(nanaimo_values)
print(kamloops_values)
print(chilliwack_values)
```

We can visualize these values in a bar graph.

```python
# Setup variables for creating the bar graph
x = np.arange(len(public_transit))  # Create x-axis positions
width = 0.20  # bar width to fit four cities side by side

# Create a bar graph comparing each column between Langley, Nanaimo, and Kamloops
plt.figure(figsize=(12, 8))  # Adjust figure size
plt.bar(x - 1.5 * width, langley_values, width, label='Langley', color='red')
plt.bar(x - 0.5 * width, nanaimo_values, width, label='Nanaimo', color='blue')
plt.bar(x + 0.5 * width, kamloops_values, width, label='Kamloops', color='green')
plt.bar(x + 1.5 * width, chilliwack_values, width, label='Chilliwack', color='purple')

# Add labels to the bar graph
plt.xlabel('Commute Time Periods (Minutes)', fontsize=12)
plt.ylabel('Number of Commuters', fontsize=12)
plt.title('Comparison of Public Transit Commuting Times', fontsize=14)
plt.xticks(x, pretty_public_transit, rotation=60, fontsize=10)  # Update x-ticks if necessary
plt.legend(fontsize=12)
plt.tight_layout()
plt.show()
```

From this graph, we can see that the majority of Langley commuters reliant on public transit are taking anywhere from 15-45 minutes for their daily commute, similar to the cities with the most efficient public transit commuting duration. Howeverk, in Langley, there is a staggering and disproportionate amount of commuters that spend over 60 minutes on transit for their daily commute duration. This statistic is an outlier when compared to the other cities, showing that it is a significant issue that only Langley faces, which may be due to lacking availability or efficiency of public transit throughout the city.

```python
# Load the GTFS zip files
translink_feed = gk.read_feed("../data/raw/translink_data.zip", dist_units="km")
chilliwack_feed = gk.read_feed("../data/raw/chilliwack_data.zip", dist_units="km")
nanaimo_feed = gk.read_feed("../data/raw/nanaimo_data.zip", dist_units="km")
kamloops_feed = gk.read_feed("../data/raw/kamloops_data.zip", dist_units="km")
```

Langley is part of the lower mainland, and the native public transit system is called "TransLink". TransLink's GTFS data includes data from all cities within the lower mainland, so we need to clean the data first to focus on transit data for Langley specifically.

Langley has two major bus loops, "Langley Centre" and "Carvolth Exchange". All bus routes that pass through Langley either start, end, or make a stop at either or both bus loops. We can filter the TransLink GTFS data, which contains all GTFS data for the Lower Mainland, so we can focus on GTFS data specific to Langley.

```python
# Get stop IDs from trips associated with Langley routes
langley_route_ids = translink_feed.routes[
    translink_feed.routes['route_long_name'].str.contains("Langley", case=False, na=False)
]['route_id'].unique()

print(langley_route_ids)
```

```python
# Filters trips DataFrame to find rows matching langley_route_ids
langley_trip_ids = translink_feed.trips[
    translink_feed.trips['route_id'].isin(langley_route_ids)
]['trip_id'].unique()
print(langley_trip_ids)
```

```python
# Count the number of trips
langley_trip_count = len(langley_trip_ids)
print(f"The number of trips taken to/from Langley Centre or Carvolth Exchange was: {langley_trip_count}")
```

Now we can repeat the process with the rest of the cities (Chilliwack, Nanaimo, Kamloops). These cities all use BC Transit, which means we can make a function to quickly:
1. get the route IDs for each city
2. get the trip IDs associated to each route ID
3. count the number of trip IDs to find the number of trips for each city

```python
def calculate_city_trip_count(feed, city_route_names):
    """
    Calculates the total number of trips for a given city based on its route names.

    Parameters:
        feed (gtfs_kit.Feed): The GTFS feed for the city.
        city_route_names (list): A list of route names for the city.

    Returns:
        int: Total number of trips for the city.
    """
    # Step 1: Get route IDs for the city's routes
    city_route_ids = feed.routes[
        feed.routes['route_long_name'].isin(city_route_names)
    ]['route_id'].unique()
    
    # Step 2: Get trip IDs for these route IDs
    city_trip_ids = feed.trips[
        feed.trips['route_id'].isin(city_route_ids)
    ]['trip_id'].unique()
    
    # Step 3: Count the trips
    city_trip_count = len(city_trip_ids)
    return city_trip_count
```

```python
# Nanaimo Routes
nanaimo_routes = [
    "Nanaimo Rapid Line", "Fairview", "Harewood", "Cinnabar", "Cedar", "Intercity",
    "Lantzville", "VIU Connector via Parkway", "VIU Connector via Jingle Pot", 
    "Hammond Bay", "Hammond Bay via Dover", "Ferry Shuttle", "NRGH", "VIU Express",
    "Downtown/Woodgrove", "NCX Nanaimo - Cowichan Express", "Downtown/Country Club",
    "Cassidy/Airport", "Parksville", "Intercity", "Hammond Bay", "Eaglecrest", "Qualicum Beach"
    "Deep Bay"
]

# Calculate Nanaimo's trip count
nanaimo_trip_count = calculate_city_trip_count(nanaimo_feed, nanaimo_routes)
print(f"The number of trips taken daily in Nanaimo is: {nanaimo_trip_count}")
```

```python
# Kamloops Routes
kamloops_routes = [
    "Tranquille", "Parkcrest", "Westsyde", "Pacific Way", "Pineview", "Downtown Loop",
    "Aberdeen", "Gleneagles", "North Shore TRU Express", "Yellowhead", "Batchelor Heights", 
    "Juniper Ridge", "Dallas", "Mt. Paul", "Pacific Way - School Special", 
    "Sahali Express - School Special", "South Kamloops - School Special",
    "South Kamloops/Aberdeen - School Special", "Pacific Way - School Special",
    "Sahali Secondary - School Special", "Landsdowne - School Special",
    "Aberdeen - School Special", "SKS Express - School Special", "Sahali/Aberdeen - School Special",
    "Batchelor Heights - School Special", "Westsyde/Batchelor Heights - School Special"
]

# Calculate Kamloops's trip count
kamloops_trip_count = calculate_city_trip_count(kamloops_feed, kamloops_routes)
print(f"The number of trips taken daily in Kamloops is: {kamloops_trip_count}")
```

```python
# Chilliwack Routes
chilliwack_routes = [
    "Highstreet/UFV", "Highstreet/McMillan", "Clearbrook/Huntingdon", "Saddle/Montrose",
    "Clearbrook/Bourquin via Hospital", "Gladwin", "Sumas Mountain", "UFV Connector",
    "Auguston/Montrose", "McMillan/Montrose", "Industrial", "Fraser Hwy",
    "Valley Connector", "West Heights", "Cedar Valley", "East Side", "Hatzic",
    "Vedder", "Evans", "Fairfield", "Promontory", "Yarrow - Greendale",
    "Broadway", "Tyson", "Fraser Valley Express", "Agassiz-Harrison", "Hope",
    "East Townline School Special", "Centre Loop School Special",
    "Sandy Hill/Bourquin School Special", "Sahali Secondary - School Special",
    "Pacific Way - School Special", "Sahali Express - School Special",
    "Batchelor Heights - School Special", "Yellowhead - School Special",
    "Juniper Ridge - School Special"
]

# Calculate Chilliwack's trip count
chilliwack_trip_count = calculate_city_trip_count(chilliwack_feed, chilliwack_routes)
print(f"The number of trips taken daily in Chilliwack is: {chilliwack_trip_count}")
```

We can now graph the daily frequency of bus routes throughout each city to easily compare Langley to the top performing cities.

```python
# Data for the bar chart
cities = ["Langley", "Chilliwack", "Nanaimo", "Kamloops"]
frequencies = [langley_trip_count, chilliwack_trip_count, nanaimo_trip_count, kamloops_trip_count]

# Create the bar chart
plt.figure(figsize=(10, 6))
plt.bar(cities, frequencies, color=["red", "blue", "green", "purple"])
plt.title("Total Trips per City", fontsize=16)
plt.xlabel("City", fontsize=14)
plt.ylabel("Total Daily Frequency (Number of Trips)", fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.show()
```

Compared to the other top-performing cities, Langley actually has more trips coming and going from the major bus exchanges within the city. So, why is their transit times not the same, if not better, than Chilliwack, Nanaimo, or Kamloops?

We can compare the duration of trips to the population of each Geography to see the relative efficiency of the transit systems across all four cities.

```python
# From "04_eda":
langley_population = 132603.0
chilliwack_population = 93203.0
nanaimo_population = 99863.0
kamloops_population = 90280.0
```

We can see that Langley has the highest population out of all of these cities.

```python
# From "05_automobile_case_study"
langley_area_km2 = 11.80
chilliwack_area_km2 = 261.34
nanaimo_area_km2 = 90.45
kamloops_area_km2 = 297.93
```

We can also see that Langley has the smallest area (in kilometres squared) out of all of these cities. If we calculate the population density per kilometres squared, Langley will far surpass the other cities.

```python
# Calculate the population density of each city
langley_pop_density = langley_population / langley_area_km2
chilliwack_pop_density = chilliwack_population / chilliwack_area_km2
nanaimo_pop_density = nanaimo_population / nanaimo_area_km2
kamloops_pop_density = kamloops_population / kamloops_area_km2

population_densities = [langley_pop_density, chilliwack_pop_density, nanaimo_pop_density, kamloops_pop_density]

# Graph the population density of each city
plt.figure(figsize=(10, 6))
plt.bar(cities, population_densities, color=["red", "blue", "green", "purple"])
plt.title("Population Density per City", fontsize=16)
plt.xlabel("City", fontsize=14)
plt.ylabel("Population Density (residents per kilometre squared)", fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.show()
```
