# Transportation Method Case Study

## Automobile Case Study

In "05_automobile_case_study", I took the top two cities, Chilliwack, Nanaimo, and compared them to Langley. The first part of this process was to compare the commuting times for both automobile-related transportation methods by time period.

### Figure 6. Comparison of Car/Truck/Van Commuting Times



From this visualization, we can see that the leading cities, Chilliwack and Nanaimo, managed to keep commuting times under 30 minutes for over 85% of their populations, with the majority completing their commute in under 15 minutes. 

In Langley, only 50% of commuters were spending less than 30 minutes on their commute, with more commuters spending 15-30 minutes than less than 15 miutes. Most Langley commuters were on the road for more than 15 minutes, with more individuals falling into the 30-45, 45-60, or >60 categories in Langley than in either other city.

### Figure 7. Comparison of Motorcycle/Scooter/Moped Commuting Times



Similar to Figure 6, we can see that in Chilliwack and Nanaimo, most commuters are spending less than 15 minutes on their daily commute, while in Langley, most commuters are spending over 15 minutes.

### Figure 8. Road Network Visualization



Next, we were able to use a shapefile from 2021 to visualize the road networks in Langley in comparison to Nanaimo and Chilliwack. Immediately, viewers can tell that Langley has a much more "grid-shaped" road network, typical of sub-urban or urban ciites. Contrastingly, Nanaimo and Chilliwack have very loose, seemingly random road networks that are typical of larger, rural areas.

To confirm this, we calculated and graphed the road density of each city.

### Figure 9. Road Density Comparison



Lastly, we broke down the road structure by counting how much (in kilometres) of each type of road there was for each city. In Langley, the majority of road was AVE and ST - avenues and streets. Avenues and streets are used in sub-urban cities to differentiate direction and provide locational clarity. In BC, streets run "vertically" (from north to south), while avenues run "horizontally" (from west to east). Avenues are enumerated, and as the number gets closer to 0, it indicates your location is getting closer to the border of the United States of America.

Meanwhile, Nanaimo and Chilliwack have a lot more "RD", "DR", and "HWY" - road, drive, highway. 'Roads' and drives are typically smaller-scale roads that are built off of larger roads to provide access to houses. Highways are present in Langley as well, however there is a lot more highway in Nanaimo and Chilliwack, which may indicate that the cities have a high dependency on highways for major roadways compared to Langley, which depends on a complex grid-like street structure.


## Public Transit Case Study

In "06_public_transit_case_study", I compared Nanaimo, Chilliwack, and Kamloops's public-transit data to Langley's public-transit related data. This case study started very similar to the last, as I graphed the time periods for each city for public transit commuters.

### Figure 10. Comparison of Public Transit Commuting Times



From this graph, we can see that Langley does seem to have a very similar proportion of public-transit commuters with similar commuting durations to the three top performing cities, however we can also see a large proportion of commuters that reported a commuting time of over an hour, completely disproportionate to the other cities and the general trends of the rest of the graph.

To analyze this further, I downloaded the General Transit Feed Specification (GTFS) data from TransLink (for the lower mainland, including Langley) and BC Transit (for Nanaimo, Chilliwack, and Kamloops). After downloading this data, I was able to find how many unique trips were made daily from bus exchanges within each region. This would indicate how many bus trips were made into or out of the city each day.

### Figure 11. Total Trips per City



From this graph, we can see that Langley actually makes the most bus trips out of any of the cities, more than all three top-performing cities. To investigate this further, I decided to compare the population density of each city. This would reveal whether there were more commuters within a smaller area, which could be a main source of the commuting issue that Langley is currently facing.

### Figure 12. Population Density per City



This was staggering. Langley's overwhelming population density meant that more individuals were commuting within a city that was multiple times smaller than the "top performing cities". Lastly, I decided to normalize the amount of trips per day, per city. If Langley's normalized trips per day was similar to the other top performing cities, it would suggest that the public transit-specific infrastructure (such as lacking busses) isn't the main issue for the city, but rather pure overpopulation of the city and the shared issue of road infrastructure (shared between public transit and automobile transportation methods) were the main issues.

# Conclusion

Langley's commuting issues stem from lacking commuting-infrastructure - that being adequately sized roads, availability of major roads and alternative routes, as well as efficiency of major routes.
